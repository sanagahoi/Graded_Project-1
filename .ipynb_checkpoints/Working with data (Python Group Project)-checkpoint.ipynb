### 4. Convert the columns "Value", "Wage", "Release Clause" to float datatype
after getting rid of currency symbol and suffix. (6 points)
- Note: When the record/entry has "M"(indicates millions) as suffix you
need to multiply that value with 1000000
- When the record/entry has "K"(indicates thousands) as suffix you
need to multiply that value with 1000 

```python
    def convert_currency1(x):
        if isinstance(x, str):
            return (x.replace('M', '').replace(',', ''))
        return (x)
```

```python
    def convert_currency2(x):
        if isinstance(x, str):
            return (x.replace('€', '').replace(',', ''))
        return (x)
```

```python
    def convert_currency3(x):
        if isinstance(x, str):
            return (x.replace('K', '').replace(',', ''))
        return (x)
```

```python
df1['Value'] = df1['Value'].apply(convert_currency1)
df1['Value'] = df1['Value'].apply(convert_currency2)
df1['Value'] = df1['Value'].apply(convert_currency3)
df1['Value'] = df1['Value'].astype('float')
```

```python
df1['Wage'] = df1['Wage'].apply(convert_currency1)
df1['Wage'] = df1['Wage'].apply(convert_currency2)
df1['Wage'] = df1['Wage'].apply(convert_currency3)
df1['Wage'] = df1['Wage'].astype('float')
```

```python
df1['Release Clause'] = df1['Release Clause'].apply(convert_currency1)
df1['Release Clause'] = df1['Release Clause'].apply(convert_currency2)
df1['Release Clause'] = df1['Release Clause'].apply(convert_currency3)
df1['Release Clause'] = df1['Release Clause'].astype('float')
```

```python
df1['Value'] = df1['Value']*1000000
df1['Wage'] = df1['Wage']*1000
df1['Release Clause'] = df1['Release Clause']*1000000
```

### 7. The column 'Height' is in inches with a quotation mark, Convert to float with decimal points. (2 points)

```python
#in centimeter
def height_conversion(height):
    if((pd.isna(height))!= True):
        chk = str(height)
        h = []
        h = chk.split("'")  
        ft = float(h[0])
        if( h[1] != ''):
            inch = float(h[1])
        else:
            inch = 0
        tot_inc = inch + ft*12
        h = tot_inc * 2.54
        return h
    else:
        return height
    
df1['Height'] = df1['Height'].apply(height_conversion)
```

```python
df1['Height'] = df1['Height'].fillna(value = df1['Height'].mean())
df1['Height'] = df1['Height'].round(2)
df1.head()
```

### 8. The column "Weight" has the suffix as lbs, remove the suffix and convert to float. (2 points)

```python
#conversion in kg
def weight_conversion(weight):
    
    if((pd.isna(weight))!= True):
        w = int(weight[0:-3])*0.453592
        return w
    else:
        return weight
    
df1['Weight'] = df1['Weight'].apply(weight_conversion)
```

```python
df1['Weight'] = df1['Weight'].fillna(value = df1['Weight'].mean())

df1.head()
```

```python
# Rounding the float value in 2 decimal no.
df1['Weight'] = df1['Weight'].round(2)
df1.head()
```
