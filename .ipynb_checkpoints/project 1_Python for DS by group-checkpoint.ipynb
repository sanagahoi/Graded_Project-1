<strong>‚óè Data Cleaning and Preprocessing

4. Convert the columns "Value", "Wage", "Release Clause" to float datatype
after getting rid of currency symbol and suffix. 
- Note: When the record/entry has "M"(indicates millions) as suffix you
need to multiply that value with 1000000
- When the record/entry has "K"(indicates thousands) as suffix you
need to multiply that value with 1000</strong>

```python
## Unique values and field to be dropped

print("Unique vaules in club logo,flag and Photo")
print(Fifa_Df['Club Logo'].nunique(),Fifa_Df['Flag'].nunique(),Fifa_Df['Photo'].nunique())
```

```python
## Drop redundant columns
# Fifa_Df.drop('Club Logo',axis=1,inplace=True)
# Fifa_Df.drop('Photo',axis=1,inplace=True)
# Fifa_Df.drop('Flag',axis=1,inplace=True)
# Fifa_Df.drop('Loaned From',axis=1,inplace=True)
Fifa_Clean=Fifa_Df.drop(columns=['Club Logo','Photo','Flag','Loaned From'],axis=1,inplace=True)
```

```python
Fifa_Df.head(3)
```

```python
# Getting rid of currency symbol and suffix.
Fifa_Df['Value']=Fifa_Df['Value'].str.replace("‚Ç¨","")
Fifa_Df['Wage']=Fifa_Df['Wage'].str.replace("‚Ç¨","")
Fifa_Df['Release Clause']=Fifa_Df['Release Clause'].str.replace("‚Ç¨","")
```

```python
Fifa_Df.info()
```

```python
#Function to replace "M" to 1000000 , "K" to 1000 and replace empty values to 0
def varb_clear(arg):
    a=1
    if pd.isnull(arg) == True:
        arg = 0
        a = 0
    else:
        if 'M' in arg:
            a = 1000000
        elif 'K' in arg:
            a = 1000
        else:
            a = 1
        arg = arg.replace('M','').replace('K','')

    return(float(arg) * a)

```

```python
#Remove NaN values in columns such as 'Value' 'Wage' 'Release Clause'
Fifa_Df.fillna({'Value':0},inplace=True)
Fifa_Df.fillna({'Wage':0},inplace=True)
Fifa_Df.fillna({'Release Clause':0},inplace=True)
```

```python
Fifa_Df['Release Clause'].dtype
```

```python
#Calling function varb_clear() to remove 'M' ,'K' in replace digits
Fifa_Df['Value']=Fifa_Df.apply(lambda l:varb_clear(l['Value']),axis=1)
Fifa_Df['Wage']=Fifa_Df.apply(lambda a:varb_clear(a['Wage']),axis=1)
Fifa_Df['Release Clause']=Fifa_Df.apply(lambda m:varb_clear(m['Release Clause']), axis=1)
```

```python
Fifa_Df.head(3)
```

```python
pd.set_option('display.max_columns', None)
```

<strong>5. Convert the column "Joined" into integer data type with keeping only the
year.</strong>

```python
#Convert the column "Joined" into integer data type with keeping only the year
Fifa_Df.fillna({'Joined':'0000'}, inplace=True)
Fifa_Df['Joined'] = Fifa_Df['Joined'].str[-4:].astype(int)
Fifa_Df.head()
```

<strong>6. Convert the column "Contract Valid Until" to pandas datetime type.</strong>

```python
# Convert the column "Contract Valid Until" to pandas datetime type
contract=Fifa_Df['Contract Valid Until']
contract=np.where(contract.isnull(),np.nan, ('12/31/' + contract))
Fifa_Df['Contract Valid Until']=pd.to_datetime(contract,errors='coerce')
Fifa_Df.head(3)
```

<strong>18. Find the position in the pitch where the maximum number of players play
and the position where the minimum number of players play? Display it using a plot.</strong>

```python

Df_pos=Fifa_Df.groupby(['Position'])['Position'].count().reset_index(name='Count').sort_values(by=('Count'),ascending=False)
print("Maximum players paying at Position :"+Df_pos['Position'][Df_pos['Count']==Df_pos['Count'].max()].iloc[0])
print("Minimum players paying at Position :"+ Df_pos['Position'][Df_pos['Count']==Df_pos['Count'].min()].iloc[0])
```

```python
# Display it using a plot.
plt.figure(figsize=(20,10))
sns.lineplot(x= 'Position', y= 'Count', data = Df_pos, color='red')
plt.show()
```

Ploting position and number of player
